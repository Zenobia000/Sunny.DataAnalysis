## Exercise 1

_This exercise is based on the Ex1data.csv dataset_

2.1 Split data into a training set data.tr and a testing set data.te. The 75% of dataset should be used of training, the remaining are used for testing. In each of the following points, the models are to be fit to data.tr, the accuracy is to be analysed on both data.tr and data.te.

```python
#Load data
data = pd.read_csv('Ex1data.csv',index_col=0)

#Initialise random seed for reproducability
random_state = 100000
np.random.seed(random_state)

#Split data into training and testing data
mask = np.random.rand(len(data)) < 0.75
```

```python
train = data[mask]
test = data[~mask]

X_train = train.iloc[:,1:]
y_train = train.iloc[:,0]
X_test = test.iloc[:,1:]
y_test = test.iloc[:,0]
```

 2.2 Train a cross-validated lasso regression. Compute the sum of squares of error for both the training and testing set.

```python
from sklearn.linear_model import LassoCV
from sklearn.metrics import mean_squared_error

model = LassoCV(cv=10)
model.fit(X_train, y_train)

print('Train error cv Lasso =',
      len(train)*mean_squared_error(model.predict(X_train), y_train))
print('Test error cv Lasso =',
      len(test)*mean_squared_error(model.predict(X_test), y_test))
```

_For comparison, lets train a normal (unrestricted) linear model:_

```python
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
model1.fit(X_train, y_train)

print('Train error unrestricted linear model=',
      len(train)*mean_squared_error(model1.predict(X_train), y_train))
print('Test error unrestricted linear model=',
      len(test)*mean_squared_error(model1.predict(X_test), y_test))
```

_-> As we see, training error is lower, but test error much higher. Typical sign for overfitting._

2.3 Given the best fit models, list the relevant variables (the ones with non-zero coefficients). Run a simple linear regression using these input variables only. Compare the sum of squares of errors of this model to the best fit model of the previous part.

```python
ind = np.nonzero(model.coef_)[0]
print(data.columns.values[ind])

model2 = LinearRegression()
model2.fit(X_train.iloc[:,ind], y_train)

print('Train error restricted linear model=',
      len(train)*mean_squared_error(model2.predict(X_train.iloc[:,ind]), y_train))
print('Test error restricted linear model=',
      len(test)*mean_squared_error(model2.predict(X_test.iloc[:,ind]), y_test))
```

2.4 Train the cross-validated Lasso model once with standardisation and once without it. Investigate the accuracy of the fits, and compare the lists relevant variables of the two models. Interpret the outcome.

```python
model3 = LassoCV(cv=10,normalize=True)
model3.fit(X_train, y_train)

print('Train error cv Lasso with standardised data =',
      len(train)*mean_squared_error(model3.predict(X_train), y_train))
print('Test error cv Lasso with standardised data =',
      len(test)*mean_squared_error(model3.predict(X_test), y_test))

ind3 = np.nonzero(model3.coef_)[0]
print(data.columns.values[ind3])
```

-> _The standardised version scales the variables to similar level, hence the decision on which ones to keep and which ones to drop does not depend on the differences in the order of magnitudes of the original variables._

2.5 Train the cross-validated ridge-regression model, investigate the value of coefficients for the whole range of the penalisation parameter, investigate the accuracy, display and compare the results.

```python
from sklearn.linear_model import RidgeCV

model4 = RidgeCV(cv=10)
model4.fit(X_train, y_train)

print('Train error cv Ridge =',
      len(train)*mean_squared_error(model4.predict(X_train), y_train))
print('Test error cv Ridge =',
      len(test)*mean_squared_error(model4.predict(X_test), y_test))
```

2.6 Plot figures showing how the selected coefficients for both lasso and ridge regressions vary when penalisation parameter Lambda is increasing. Interpret the plots by comparing the shrinkage effect of Lambda for both models.

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge

lassos = Lasso()
ridges = Ridge()

coefs_l = []
errors_l = []

coefs_r = []
errors_r = []

alphas = np.logspace(-2, 5, 200)

# Train the modelS with different regularisation strengths
for a in alphas:
    
    lassos.set_params(alpha=a)
    lassos.fit(X_train, y_train)
    coefs_l.append(lassos.coef_)
    
    ridges.set_params(alpha=a)
    ridges.fit(X_train, y_train)
    coefs_r.append(ridges.coef_)

        
# Display results
plt.figure(figsize=(20, 20))

plt.subplot(211)
ax = plt.gca()
ax.plot(alphas, coefs_l)
ax.set_xscale('log')
plt.xlabel('lambda')
plt.ylabel('coefficients')
plt.title('Lasso coefficients as a function of the regularization')
plt.axis('tight')


plt.subplot(212)
ax = plt.gca()
ax.plot(alphas, coefs_r)
ax.set_xscale('log')
plt.xlabel('lambda')
plt.ylabel('coefficients')
plt.title('Ridge coefficients as a function of the regularization')
plt.axis('tight')

plt.show()
```

-> _The plot of coefficients shows that Lasso actually properly eliminates variables when the penalisation is strong enough, whereas Ridge regression tends to use most of the variables with reduced coefficients._

## Exercise 2

_This exercise is based on the dataset provided on the problem sheet, which is to be read in from txt file_

```python
file = open('house.data.txt') 
firstline = file.readline()
f, n = firstline.split()
f = int(f)
n = int(n)
data = file.readlines()
clf = LinearRegression() 
x_train = []
y_train = []
for x in data[:n]:
    tmp = x.split()
    tmp = [float(f) for f in tmp] 
    x_train.append(tmp[0: len(tmp) - 1]) 
    y_train.append(tmp[len(tmp) - 1])
clf.fit(x_train, y_train)

x_test = []
for x in data[n+1:]:
    tmp = x.split()
    tmp = [float(f) for f in tmp] 
    x_test.append(tmp)

y_test = clf.predict(x_test) 
for y in y_test:
    print(y)
```
